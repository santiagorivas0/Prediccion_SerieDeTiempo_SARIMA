```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('Merge_santi.csv')
df.head()
```

```
print(df.columns)
```

```
# convertir columna de fecha y hora a formato de fecha y hora
df['Dias_acumulados'] = pd.to_datetime(df['Dias_acumulados'], format='%Y-%m-%d %H:%M:%S')

# sumar 293 días a la columna de días acumulados
df['Dias_acumulados'] += pd.Timedelta(days=293)

# establecer la columna de fecha y hora como el índice del DataFrame
df.set_index('Dias_acumulados', inplace=True)

df
```

```
# trazar una serie de tiempo con etiquetas de eje y título
df.plot()
plt.xlabel('Días Acumulados')
plt.ylabel('Valor en Reales')
plt.title('Serie de tiempo')
plt.show()
```

```
# calcular la media móvil de 365 días
rolling_mean = df.rolling(window=365).mean()

# trazar la serie de tiempo y la media móvil
plt.plot(df)
plt.plot(rolling_mean, label='Media móvil de 365 días')
plt.legend()
plt.xlabel('Días Acumulados')
plt.ylabel('Valor en Reales')
plt.title('Serie de Tiempo con Media Móvil')
plt.show()
```
